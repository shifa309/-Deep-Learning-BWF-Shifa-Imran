```python
import pandas as pd
import glob

# Define the path where the csv files are located
path = r'C:\Users\DELL\Desktop\-Deep-Learning-BWF-Shifa-Imran\MiniProject1\US_Census Data Cleaning'

# Use glob to get all the csv files in the path
all_files = glob.glob(path + "/*.csv")

dfs = []

# Loop through each csv file and load it into a dataframe
for filename in all_files:
    df = pd.read_csv(filename, header=0)
    dfs.append(df)

# Concatenate all the dataframes into one
us_census = pd.concat(dfs, axis=0, ignore_index=True)

# Save the merged dataframe as a new Excel file
#us_census.to_excel('us_census.xlsx', index=False)

```

```python
# Print the columns and data types of the merged dataframe
print("Columns:")
print(us_census.columns)
print("\nData Types: ")
print(us_census.dtypes)


#the data types of the columns in the us_census dataframe might hinder us as we try to make histograms, 
#depending on the specific columns we are interested in. For example, some columns might be stored as strings 
#which would make it difficult to plot them as numerical values. 
#Similarly, some columns might have missing values represented as strings (e.g. "NaN"), 
#which would need to be converted to NaN values before plotting.


# Select the non-numerical columns and print their names
non_numeric_columns = us_census.select_dtypes(exclude=['int64', 'float64']).columns.tolist()
print("\nNon Numeric Columns: ")
print(non_numeric_columns)

#- Look at the .head() of the DataFrame 
#so that you can understand why some of these dtypes are objects instead of integers or floats.

print("\nPrint the first five rows of the merged dataframe using .head()")
print(us_census.head())
```

```python
# Fill any missing values in the 'GenderPop' column with '0'
us_census['GenderPop'].fillna(value='0', inplace=True)

# Split the GenderPop column into Men and Women columns
us_census[['Men', 'Women']] = us_census['GenderPop'].str.split('_', expand=True)

# Remove non-numeric characters and the M/F character from the Men and Women columns
us_census['Men'] = us_census['Men'].replace('[^0-9]', '', regex=True).astype(int)

# Remove non-numeric characters and convert to int
us_census['Women'] = us_census['Women'].replace('[^0-9]', '', regex=True)

#there were some rows for female that has no numeric value but _F only so for that following need to be done
#The errors='coerce' parameter in to_numeric() ensures that any non-numeric values are converted to NaN 
#The fillna(0) method replaces the NaN values with 0.
us_census['Women'] = pd.to_numeric(us_census['Women'], errors='coerce')
us_census['Women'] = us_census['Women'].fillna(0).astype(int)


# Remove the M/F character from the Men and Women columns
us_census['Men'] = us_census['Men'].replace('M', '', regex=True).astype(int)
us_census['Women'] = us_census['Women'].replace('F', '', regex=True).astype(int)

# Drop the original GenderPop column
us_census = us_census.drop(columns=['GenderPop'])

#us_census.to_excel('us_census_modified.xlsx', index=False)

```

```python
import matplotlib.pyplot as plt

#Did you get an error? These monstrous csv files probably have nan or null values in them! Print out your column with the number of women per state to see.
#- We can fill in those nans by using pandas’ .fillna() function.
#I ALREADY USED .FILLNA() before due to error encountereed before so doesn't need it now.

# Make scatterplot
plt.scatter(us_census['Women'], us_census['Income'])
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
import numpy as np

#as i already converted to 0 so here converting it again to NaN 
#to apply the follwoing condition for filling NaN vaues in subsequent lines of code

# Convert 0 to NaN in Women column
us_census['Women'] = us_census['Women'].replace(0, np.nan)

# Fill NaN values in Women column
us_census['Women'] = us_census['Women'].fillna(us_census['TotalPop'] - us_census['Men'])
print(us_census['Women'])
```

```python
# check for duplicates
print(us_census.duplicated().value_counts())

# drop duplicates
us_census.drop_duplicates(inplace=True)

# make scatterplot
plt.scatter(us_census['Women'], us_census['Income'])
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
#histograms

import matplotlib.pyplot as plt

# Convert columns to numerical format
us_census['Hispanic'] = us_census['Hispanic'].replace('[%]', '', regex=True).astype(float)
us_census['White'] = us_census['White'].replace('[%]', '', regex=True).astype(float)
us_census['Black'] = us_census['Black'].replace('[%]', '', regex=True).astype(float)
us_census['Native'] = us_census['Native'].replace('[%]', '', regex=True).astype(float)
us_census['Asian'] = us_census['Asian'].replace('[%]', '', regex=True).astype(float)
us_census['Pacific'] = us_census['Pacific'].replace('[%]', '', regex=True).astype(float)

# Fill nan values with the mean of the column
us_census = us_census.fillna(value={'Hispanic':us_census['Hispanic'].mean(),
                                    'White':us_census['White'].mean(),
                                    'Black':us_census['Black'].mean(),
                                    'Native':us_census['Native'].mean(),
                                    'Asian':us_census['Asian'].mean(),
                                    'Pacific':us_census['Pacific'].mean()})

# Check for duplicates
duplicates = us_census.duplicated()
print(duplicates.value_counts())

# Create histograms
plt.hist(us_census['Hispanic'], bins=10)
plt.title('Hispanic Population Percentage')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()

plt.hist(us_census['White'], bins=10)
plt.title('White Population Percentage')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()

plt.hist(us_census['Black'], bins=10)
plt.title('Black Population Percentage')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()

plt.hist(us_census['Native'], bins=10)
plt.title('Native Population Percentage')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()

plt.hist(us_census['Asian'], bins=10)
plt.title('Asian Population Percentage')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()

plt.hist(us_census['Pacific'], bins=10)
plt.title('Pacific Population Percentage')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()
```
